```python
import json

def filter_json(input_file, output_file):
    # Read the input JSON file
    with open(input_file, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # Filter the data with a step of 2
    filtered_data = data[::24]

    # Write the filtered data to the output JSON file
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(filtered_data, f, indent=4, ensure_ascii=False)

# Example usage
input_filename = "formatted_codeforces_python_submissions.json"
output_filename = "formatted_output.json"
filter_json(input_filename, output_filename)
```

```python
import faiss
import numpy as np
import json
from sentence_transformers import SentenceTransformer
```

```python
import pandas as pd
import json

# Load merged benchmark dataset
with open("merged_benchmarks.json", "r") as f:
    benchmarks = json.load(f)

# Print overview of merged data
actual_length = len(benchmarks)
print("Data Overview:")
print(f"Total Entries: {actual_length}")
print("Sample Entries:")
for sample in benchmarks[:5]:
    print(sample)

# Save merged dataset
output_file = "merged_benchmarks.json"
with open(output_file, "w") as f:
    json.dump(benchmarks, f, indent=4)

print(f"Merged dataset saved as {output_file}")
```

```python
# Extract benchmark questions.
questions = [b["question"] for b in benchmarks]

# Initialize the embedding model using all-mpnet-base-v2.
print("Loading all-mpnet-base-v2 model and computing embeddings...")
model = SentenceTransformer('all-mpnet-base-v2')
embeddings = model.encode(questions, convert_to_numpy=True)
```

```python
# Determine the dimensionality of the embeddings.
d = embeddings.shape[1]
print(f"Embedding dimension: {d}")
```

```python
# Create a FAISS index using L2 distance.
print("Creating FAISS index...")
index = faiss.IndexFlatL2(d)
index.add(embeddings)
print(f"Added {index.ntotal} embeddings to the index.")
```

```python
# Save the FAISS index to disk.
index_file = "benchmark.index"
faiss.write_index(index, index_file)
print(f"FAISS index saved to {index_file}")
```

```python
# Save the benchmark metadata to disk.
metadata_file = "benchmarks.json"
with open(metadata_file, "w") as f:
    json.dump(benchmarks, f, indent=2)
print(f"Benchmark metadata saved to {metadata_file}")
```

```python
# prompt: can i visualize the vector database? and use different colors for each benchmark (use 3 PCAs instead of 2 ) (use plotly for a full 3D rotatable plot)

import plotly.graph_objects as go
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# Load the embeddings and metadata
index_file = "benchmark.index"
metadata_file = "benchmarks.json"

index = faiss.read_index(index_file)
embeddings = np.array(index.reconstruct_n(0, index.ntotal))

with open(metadata_file, 'r') as f:
    benchmarks = json.load(f)

# Create a Pandas DataFrame for easier handling
df = pd.DataFrame(embeddings)
df['benchmark'] = [b['benchmark'] for b in benchmarks]


# Apply PCA to reduce dimensionality to 3
pca = PCA(n_components=3)
pca_result = pca.fit_transform(df.drop('benchmark', axis=1))


# Create the 3D scatter plot using Plotly
fig = go.Figure()

benchmark_colors = {
    "gpqa_diamond": "blue",
    "simpleqa": "red",
    "swe-bench_verified": "green",
    "codeforces_python_submissions": "purple",
    "aime_2024": "orange",
    "math_500": "brown"
}

for benchmark in df['benchmark'].unique():
    subset = df[df['benchmark'] == benchmark]
    fig.add_trace(go.Scatter3d(
        x=pca_result[df['benchmark'] == benchmark, 0],
        y=pca_result[df['benchmark'] == benchmark, 1],
        z=pca_result[df['benchmark'] == benchmark, 2],
        mode='markers',
        marker=dict(size=5, color=benchmark_colors.get(benchmark, 'gray')),  # Default to gray if color not found
        name=benchmark
    ))

fig.update_layout(
    title="Vector Database Visualization (3D PCA)",
    scene=dict(
        xaxis_title="PC1",
        yaxis_title="PC2",
        zaxis_title="PC3"
    )
)

fig.show()
```

```python
# Load the FAISS index and metadata.
print("Loading FAISS index and metadata...")
index = faiss.read_index("benchmark.index")
with open("benchmarks.json", "r") as f:
    metadata = json.load(f)
print("Index and metadata loaded.")

# Define a function to perform a nearest-neighbor search.
def search_query(query, index, metadata, model, k=1):
    # Compute the embedding for the user query.
    query_embedding = model.encode([query], convert_to_numpy=True)

    # Perform the search: retrieve k nearest neighbors.
    distances, indices = index.search(query_embedding, k)

    # Gather results with metadata.
    results = []
    for dist, idx in zip(distances[0], indices[0]):
        if idx < len(metadata):
            result = {
                "question": metadata[idx]["question"],
                "benchmark": metadata[idx]["benchmark"],
                "distance": float(dist)
            }
            results.append(result)
    return results

# Example usage:
user_query = "Polycarp urgently needs a shovel! He comes to the shop and chooses an appropriate one. The shovel that Policarp chooses is sold for *k* burles. Assume that there is an unlimited number of such shovels in the shop.\n\nIn his pocket Polycarp has an unlimited number of \"10-burle coins\" and exactly one coin of *r* burles (1<=\u2264<=*r*<=\u2264<=9).\n\nWhat is the minimum number of shovels Polycarp has to buy so that he can pay for the purchase without any change? It is obvious that he can pay for 10 shovels without any change (by paying the requied amount of 10-burle coins and not using the coin of *r* burles). But perhaps he can buy fewer shovels and pay without any change. Note that Polycarp should buy at least one shovel."
results = search_query(user_query, index, metadata, model, k=10)

print("Search results:")
for res in results:
    print(f"Matched Benchmark Question: {res['question']}")
    print(f"belongs to benchmark: {res['benchmark']}")
    print(f"Distance: {res['distance']:.4f}")
```
